```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('D:\moti\ML\Iris.csv')
df.head()
```

```python
x= df.drop('Species', axis=1)
y= df['Species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model = GaussianNB()
model.fit(x_train, y_train)

```

```python
y_pred = model.predict(x_test)
```

```python
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))

```

```python
# Convert y_test to numeric values
y_test_numeric = pd.factorize(y_test)[0]

# Plot the scatter plot
plt.scatter(x_test['SepalLengthCm'], x_test['SepalWidthCm'], c=y_test_numeric)
plt.show()
```

```python
import seaborn as sns
conf_matrix= confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', xticklabels=df['Species'].unique(), yticklabels=df['Species'].unique())
```
